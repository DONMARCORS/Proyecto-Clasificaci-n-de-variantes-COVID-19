```python
## funcion que segmenta el genoma en secuencias de tamaño length

def split_genoma(genoma,length,genoma_name):
  g = []
  y = []
  for i in range(len(genoma)):
    if genoma[i:i+length]:
      g.append(genoma[i:i+length] ) 
      y.append(genoma_name)
    # else:
    #   g.append(genoma[i:])
    #   y.append(genoma_name)

  # if len(genoma) /length == int(len(genoma) /length):
  #   for i in range(int(len(genoma)/length)):
  #     g.append(genoma[i*length:(i+1)*length] ) 
  #     y.append(genoma_name)
  #   # g.append(genoma)
  # else:
  #   for i in range(int(len(genoma)/length)):
  #     g.append(genoma[i*length:(i+1)*length])
  #     y.append(genoma_name)
  #   g.append(genoma[(i+1)*length:])
  #   y.append(genoma_name)

  return g,y  ## devolvemos las secuencias y las etiquetas
```

```python
## aplicamos la funcion a los genomas

## le pueden mover al numero de enmedio
X_1,y_1 = split_genoma(omicron,50,'omicron')
X_2,y_2 = split_genoma(delta,50,'delta')

## juntamos seucncias en una lista y lo mismo con los nombres
X = X_1 + X_2
y = y_1 + y_2
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer

# Cargar el tokenizer y el modelo pre-entrenado
tokenizer = AutoTokenizer.from_pretrained("Isaacwu/omicron-delta-classifier-pipeline")
model = TFAutoModelForSequenceClassification.from_pretrained("Isaacwu/omicron-delta-classifier-pipeline")
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear una instancia de LabelEncoder
label_encoder = LabelEncoder()

# One hot encoder de las variantes a predecir
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
# y_encoded = label_encoder.transform(y)

y_train_encoded
```

```python
import shutil
import os
import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer

path = 'model_callback'

if os.path.exists(path):
    shutil.rmtree(path)
    print('La carpeta se ha eliminado')
else:
    print("La carpeta no existe.")

tokenizer = AutoTokenizer.from_pretrained("Isaacwu/omicron-delta-classifier-pipeline")
model = TFAutoModelForSequenceClassification.from_pretrained("Isaacwu/omicron-delta-classifier-pipeline")

X_train_tokenized = tokenizer(X_train, truncation=True, padding=True, max_length=110)
X_test_tokenized = tokenizer(X_test, truncation=True, padding=True, max_length=110)

y_train_tensor = tf.convert_to_tensor(y_train_encoded)
y_test_tensor = tf.convert_to_tensor(y_test_encoded)

train_dataset = tf.data.Dataset.from_tensor_slices((dict(X_train_tokenized), y_train_tensor))
test_dataset = tf.data.Dataset.from_tensor_slices((dict(X_test_tokenized), y_test_tensor))

batch_size = 32## este se puede mover
epochs = 50   ## esto si se puede mover

train_dataset = train_dataset.shuffle(len(X_train)).repeat(epochs)
train_dataset = train_dataset.padded_batch(batch_size, padded_shapes=({'input_ids': [None], 'attention_mask': [None]}, []))

# optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
# loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
# metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')

checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath=path,
                                                         monitor='val_loss',
                                                         save_best_only=True)
lr_callback = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',
                                                    factor=0.1,
                                                    patience=5,
                                                    min_lr=0.0001
                                                )

model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy'])

history = model.fit(train_dataset,
          validation_data=test_dataset.batch(batch_size),
          epochs=epochs,
          steps_per_epoch= 20,
          validation_steps = 5,
          callbacks=[checkpoint_callback,lr_callback])

model.evaluate(test_dataset.batch(batch_size))
```

```python
import pandas as pd
data = pd.DataFrame(history.history)
```

```python
data[['loss','val_loss']].plot()
data[['accuracy','val_accuracy']].plot()
```
